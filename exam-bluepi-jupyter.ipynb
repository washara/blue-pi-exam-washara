```python
#Create spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder \
  .appName('Jupyter BigQuery Storage')\
  .config('spark.jars', 'gs://spark-lib/bigquery/spark-bigquery-latest_2.12.jar') \
  .getOrCreate()
```

```python
#Edit variable for change dataset
project_id = "exam-de-washara"
dataset_id = "user_dataset"
table_id = "user_log"

table = "exam-de-washara:user_dataset.user_log".format(project_id=project_id, \
                                                       dataset_id=dataset_id, \
                                                       table_id=table_id) \

spark_df = spark.read \
  .format("bigquery") \
  .option("table", table) \
  .load()
```

```python
#Convert spark to pandas
df = spark_df.toPandas()
```
